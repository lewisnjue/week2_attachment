````python
from bs4 import BeautifulSoup

````

````python
html_doc = """
<html>
    <head><title>Boma Analytics Sandbox</title></head>
    <body>
        <h1 id="page-title">Featured Properties in Nairobi</h1>
        
        <div class="property-card" id="prop-1">
            <h2 class="title">3 Bedroom Apartment in Kilimani</h2>
            <span class="price">KSh 15,000,000</span>
            <a class="details-link" href="https://example.com/listings/kilimani-3bed">View Details</a>
        </div>

        <div class="property-card" id="prop-2">
            <h2 class="title">5 Bedroom Villa in Karen</h2>
            <span class="price">KSh 85,000,000</span>
            <a class="details-link" href="https://example.com/listings/karen-5bed">View Details</a>
        </div>
    </body>
</html>
"""

````

````python
# 2. "Make the Soup" - Parse the raw HTML text
soup = BeautifulSoup(html_doc, 'html.parser')
````

````python

print("--- 1. Grab a Single Element ---")
# Find the main title element
title_element = soup.find('h1', id='page-title')
print(f"Raw Tag: {title_element}")
print(f"Clean Text: {title_element.text}\n")
````

````python

print("--- 2. Extracting Attributes (Links) ---")
# Find the first link and pull its destination URL
first_link = soup.find('a', class_='details-link')
url = first_link['href']
print(f"Destination URL: {url}\n")
````

````python

print("--- 3. Loop Through Multiple Elements (The Core Pattern) ---")
# Find all property cards on the page
cards = soup.find_all('div', class_='property-card')

for card in cards:
    # Look INSIDE this specific card for details
    name = card.find('h2', class_='title').text.strip()
    price = card.find('span', class_='price').text.strip()
    
    print(f"🏡 Property: {name}")
    print(f"💰 Price:    {price}")
    print("-" * 30)
````
