```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
house_url  = "https://www.propertypro.co.ke/property-for-sale/house"
apartment_url = "https://www.propertypro.co.ke/property-for-sale/flat-apartment"
```

```python
response = requests.get(house_url)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
def get_total_pages(soup, items_per_page=20):
    target_div = soup.select_one('div.onpage-filters.d-flex.justify-content-between.align-items-center')
    div_html = target_div
    soup = BeautifulSoup(str(div_html), 'html.parser')
    
    # 1. Find the strong tag containing the total items count
    strong_tag = soup.find('strong')
    if not strong_tag:
        return 0  # Or handle error as needed
        
    # Clean the text (remove commas if numbers look like 24,300)
    total_items_text = re.sub(r'\D', '', strong_tag.text)
    total_items = int(total_items_text)
    
    # 2. Calculate the total pages using integer ceiling division
    total_pages = (total_items + items_per_page - 1) // items_per_page
    
    return total_pages
```

```python
get_total_pages(soup)
```
